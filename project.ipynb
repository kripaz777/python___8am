```python
# !pip install beautifulsoup4
# !pip install requests
# !pip install lxml
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.sharesansar.com/live-trading', 'html.parser')
soup = requests.get(url)
soup
```

```python
code = BeautifulSoup(soup.text, "lxml")
code
```

```python
table_code = code.table
table_row = table_code.find_all('tr')
table_row
```

```python
# list_data = []
# for i in table_row[1:]:
#     list_data.append(i.text.split('\n\n'))

# list_data
rows = []
for i in table_row[1:]:
    tds = []
    for j in i:
        z = j.text.replace(" ","").replace('\n','')
        if z != "":
            tds.append(z)
        
    rows.append(tds)

rows
```
